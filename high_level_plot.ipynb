```python
from ggplot import *
import matplotlib.pyplot as plt
from matplotlib.dates import YearLocator
from ggplot import datasets
```

```python
ggplot(datasets.movies, aes('year','budget')) + \
geom_line(color='red') + \
scale_x_date(breaks=date_breaks('7 years')) + \
scale_y_continuous(labels='comma')
```

```python
movie_data = datasets.movies
tick_every_n = YearLocator(7)
x = movie_data.year
y = movie_data.budget
fig, ax = plt.subplots()
ax.plot(x, y, 'red')
ax.xaxis.set_major_locator(tick_every_n)
plt.show()
```
